Electric vs. Gas Cars: Is It Cheaper to Drive an EV?

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url = "https://www.nrdc.org/stories/electric-vs-gas-cars-it-cheaper-drive-ev"
```

```python
response = requests.get(url)
response.content
```

```python
soup = BeautifulSoup(response.content ,'html.parser')
```

```python
print(soup.prettify)
```

```python
para = soup.find_all('div',class_ = 'c-wysiwyg u-clearfix')
para
#para1
```

```python
para= [i.text for i in para]

```

```python
para

```

```python
df_cleaned = pd.DataFrame([para])

# Specify the path where you want to save the CSV file
csv_file_path = "Hertz_data.csv"

# Saving the cleaned data to a CSV file
df_cleaned.to_csv(csv_file_path, index=False)
```

```python
cleaned_data = ' '.join(para).replace('\n', ' ').replace('\xa0', ' ').strip()

# Creating a DataFrame with the cleaned text
df_cleaned = pd.DataFrame([cleaned_data], columns=['Cleaned_Data'])

# Specify the path where you want to save the CSV file
csv_file_path = "Hertz__clean_data.csv"

# Saving the cleaned data to a CSV file
df_cleaned.to_csv(csv_file_path, index=False)
```
